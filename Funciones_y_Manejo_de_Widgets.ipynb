#FUNCIONES EN PYTHON
Una función en Python es un bloque de código que realiza una tarea específica y puede ser reutilizado en diferentes partes de un programa. Las funciones ayudan a organizar el código, hacerlo más modular, más fácil de mantener y reutilizar.

Para definir una función en Python, se utiliza la palabra clave def, seguida del nombre de la función y paréntesis que pueden contener parámetros. Aquí tienes un ejemplo básico:

#USO DEL DEF

imagen.png

```python
#Definimos el numbre de la función
#     |
#     |  Estas 2 variables (agumentos) entran a la funcion)
#     |   |
#   |__||___|
def suma(a,b):
    resultado = a+b
    #La función retorna el resultado a quien la llame
    return resultado

#Se llama la funcion suma,
#se pasan como argumento 5 y 6
#y retorna su valor a resut_suma
resut_suma=suma(5,6)
#Mostramos el valor por pantalla
print(resut_suma)

#Lo mismo pero mas corto
print(suma(5,6))
```

#Manejo de Widgets

Un widget es una pequeña aplicación o programa diseñado para facilitar el acceso a funciones frecuentemente usadas de un dispositivo. Los widgets se integran en el escritorio de una computadora o en la pantalla de un teléfono móvil y ofrecen información visual sin necesidad de abrir una aplicación completa.

#CREACIÓN DE UN BOTON


```python
#Importamos la librería que nos permite usar los widgets
import ipywidgets as widgets
from IPython.display import clear_output

#Definimos una funcion que se va a llamar cuando
#se presione un boton determinado
def funcion_q_se_activa_al_presionar_el_boton(b):
  #Rescatamos el valor del boton que es pasado en la variable b
  valor_del_boton = b.description
  print('PRESIONASTE EL BOTON CON VALOR',valor_del_boton)


#Creamos un boton de nombre "boton"
#Que tiene la leyenda "Presioname"
boton = widgets.Button(description='Presioname')
#Mostramos el boton en pantalla
display(boton)

#Activamos el boton, hacemos que cuando se le de
#click (boton.on_click) llame a la funcion
#funcion_q_se_activa_al_presionar_el_boton
boton.on_click(funcion_q_se_activa_al_presionar_el_boton)
```

```python
bug = 0

def Despliegue_del_menu_principal(b):
  global bug
  if( b.new == 'Agregar Inventario' and bug != 1):
    clear_output()
    menu_principal()
    bug=1

    print('Agregar Inventario')


  if(b.new == 'Consultar inventario' and bug != 2):
    clear_output()
    menu_principal()
    bug=2

    print('Consultar inventario')

  if(b.new == 'Vender' and bug != 3):
    clear_output()
    menu_principal()
    bug=3

    pagina_principal_de_venta()

def menu_principal():
  print('Bienvenido al menu principal')
  menu = widgets.Select(
      options=['Agregar Inventario', 'Consultar inventario', 'Vender','SALIR'],
      value='SALIR',
      description='QUE DESEAS',
      disabled=False
  )
  display(menu)
  menu.observe(Despliegue_del_menu_principal)


menu_principal()
```
